```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Data/Employee_dataset.csv")
print(df.head())
```

```python
sns.histplot(df['age'])
```

```python
sns.countplot(x=df['gender_full'])
```

```python
x = df['STATUS_YEAR'].value_counts()
plt.pie(x.values,
        labels=x.index,
        autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.barplot(x=df['department_name'], y=df['length_of_service'])
plt.xticks(rotation='90')
```

```python
sns.scatterplot(x=df['length_of_service'],
                y=df['age'])
```

```python

sns.countplot(x=df['STATUS_YEAR'],hue=df['STATUS'])
plt.show()
```

```python

from sklearn import datasets, decomposition
iris = datasets.load_iris()
X = iris.data
y = iris.target
pca = decomposition.PCA(n_components=2)
X = pca.fit_transform(X)
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```

```python
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True)
plt.show()
```
